```python
def simple_stemmer(text):
    ps=nltk.porter.PorterStemmer()
    text = ' '.join([ps.stem(word) for word in text.split()])
    return text
```

```python
def lemmatize(word):
    doc=nlp(word)
    lemma = doc[0].lemma_
    return lemma
```

```python
def funcion_lemantizadora(df,tex_col):
    start_time=time.time()
    df_clean=df[df[tex_col].notnull()]
    df_clean.reset_index(drop=True,inplace=True)
    pre_process_lambda = lambda text:' '.join([palabra.lemma_ for palabra in nlp2(' '.join([palabra for palabra in nltk.RegexpTokenizer(r'[a-z]+').tokenizer(str(text).lower()) if palabra not in stopwords_words]))])
    df_clean[text_col]=df_clean[tex_col].apply(pre_process_lambda)
    elapsed_time=time.time()-start_time
    print('Tiempo: ',elapsed_time)
    return df_clean
```
